```python
from gtts import gTTS
from IPython.display import Audio, display

# Set the text you want to convert to speech
text = "पायथन  एक   शक्तिशाली   और   बहुपरकारी   प्रोग्रामिंग   भाषा   है"

# Create a gTTS object with the text and desired language (e.g., 'en' for English)
tts = gTTS(text=text, lang='en')

# Save the speech as an MP3 file
tts.save("speech.mp3")

audio_file_path = "speech.mp3"
audio = Audio(audio_file_path)
display(audio)
```
